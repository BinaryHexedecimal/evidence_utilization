```python
import json

def take_subset(model):
    

    refute_file = f"../data/split_by_verdict/true/{model}_1refute_true_bullet_first_result.json"
    mix_file = f"../data/split_by_verdict/true/{model}_mix_true_bullet_first_result.json"
    out_file = f"../data/split_by_verdict/true/{model}_1refute_true_bullet_first_subset_result.json"
    

    with open(refute_file, "r", encoding="utf-8") as f:
        refute_data = json.load(f)
    
    with open(mix_file, "r", encoding="utf-8") as f:
        mix_data = json.load(f)
    

    mix_claim_ids = {item["claim_id"] for item in mix_data}
    
    print(f"Found {len(mix_claim_ids)} unique claim_ids in mix file.")
    
    subset = [
        item for item in refute_data
        if item.get("claim_id") in mix_claim_ids
    ]
    
    print(f"Subset size: {len(subset)} items")
    
    with open(out_file, "w", encoding="utf-8") as f:
        json.dump(subset, f, ensure_ascii=False, indent=2)
    
    print(f"Saved subset file to:\n{out_file}")
```

```python
models = ["llama", "mistral", "qwen"]
```

```python
for model in models:
    take_subset(model)
```
