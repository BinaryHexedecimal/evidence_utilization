```python
import json
from pathlib import Path
#import parse_bullets_score as pbs
#from pathlib import Path
```

```python

try:
    BASE_DIR = Path(__file__).resolve().parent
except NameError:
    BASE_DIR = Path.cwd()
   
bert_dir = BASE_DIR / "../eval_bullet_Bert/tmp_data"
dst_dir =  BASE_DIR  / "../data/no_split_bullet_first"

bert_prefix = "bert_score_"
bert_suffix = "_bullet_first.json"

dst_suffix = "_bullet_first_result.json"
```

```python
def merge_bert_into_result(bert_dir, dst_dir, bert_prefix, bert_suffix, dst_suffix):
    models = ["llama", "mistral", "qwen"]
    modes = ["2support", "2refute", "mix"]  
    for model in models:
        for mode in modes:
            bert_file = bert_prefix + f"{model}_{mode}" + bert_suffix
            bert_path = bert_dir / bert_file
            dst_file = f"{model}_{mode}" + dst_suffix
            dst_path = dst_dir / dst_file
            
            with open(bert_path, "r", encoding="utf-8") as f:
                data_A = json.load(f)
            
            with open(dst_path, "r", encoding="utf-8") as f:
                data_B = json.load(f)
                
            bert_by_claim_id = {}
            
            for item in data_A:
                cid = item.get("claim_id")
                if cid and "bert_scores" in item:
                    bert_by_claim_id[cid] = item["bert_scores"]
            
            missing = []
            added = 0
        
            for item in data_B:
                cid = item.get("claim_id")
                if cid in bert_by_claim_id:
                    item["bert_scores"] = bert_by_claim_id[cid]
                    added += 1
                else:
                    missing.append(cid)
            print(f"Items in file A: {len(bert_by_claim_id)}")
            print(f"Items in file B: {len(data_B)}")
            print(f"BERT scores added: {added}")
        
            if missing:
                print(f"Missing bert_scores for {len(missing)} items")
                print("Example missing claim_id:", missing[:5])
            
            assert added == len(data_B), "Mismatch: not all items received bert_scores"
            with open(dst_path, "w", encoding="utf-8") as f:
                json.dump(data_B, f, ensure_ascii=False, indent=2)
            
            print(f"Saved updated file {dst_path}")
```

```python
merge_bert_into_result(bert_dir, dst_dir, bert_prefix, bert_suffix, dst_suffix)
```
