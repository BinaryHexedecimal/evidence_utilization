```python
def generate_source_file_name_lists(models, modes, suffix):
    merged_list = []
    for model in models:
        for mode in modes:
            filename = f"{model}_{mode}" + suffix
            merged_list.append(filename)
    return merged_list
```

```python

models = ["llama", "mistral", "qwen"]

modes = ["2support", "2refute", "mix"]  

suffix = "_half_true_bullet_first_result.json"


src_file_name_lst = generate_source_file_name_lists(models, modes, suffix)
print(src_file_name_lst)


```
